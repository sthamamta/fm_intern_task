# Merge, Join and Concat

Create two dataframe (first dataframe having 50 random data and second dataframe having 40 random data). Each dataframe should have 3 columns including a common column (example: id) and 90% of data in common column should have identical data                                                                                      
Merge both the dataframe using:                                                                                   Inner                                                              
Outer                                                              
Left                                                           
Right                                                        
Evaluate the differences observed                                                      
Use join() to combine the dataframe                                                                
Use concat to combine the dataframe                                        
Evaluate the differences observed between merge, join and concat              



Merge: Merging is used for doing complex column-wise combinations of dataframes in a SQL-like way. It handles duplicate values on the joining columns or index by performing a cartesian product. 

Concatination: Concatenating is for combining more than two dataframes in either column-wise or row-wise. With concatination datasets are just sticked together along an axis-either the row axis or column axis.There is not data loss in concatination.


Join: Join offers a shorthand for a subset of the use cases of merge.





```python
import random
columns = ['Weight','Height']
df_1 = pd.DataFrame(np.random.randint(0,100,size=(50, 2)), columns = columns)
id1 = random.sample(range(70), 50)
df_1['Id'] = id1
df_1.sort_values("Id")
```

```python
columns = ['Age','Score']
df_2 = pd.DataFrame(np.random.randint(0,100,size=(40, 2)), columns = columns)

col_id = df_1['Id'].tolist()
sampled_id = sample(col_id,36)  #taking random sample of id
sampled_id= sampled_id+ random.sample(range(100,120), 4)
df_2['Id'] = sampled_id
df_2.sort_values("Id")
```

```python
# inner merge
df_inner = pd.merge(df_1,df_2,on='Id').set_index('Id').sort_values('Id')
df_inner

# df_inner.count()
# df_inner.isnull().sum(axis = 0)

#comments
# only the ids that are present in both are merged 
```

```python
df_outer= pd.merge(df_1,df_2,on='Id',how='outer').set_index('Id').sort_values('Id')
df_outer
# df_outer.count()
# df_outer.isnull().sum(axis = 0)
```

```python
df_right = pd.merge(df_1,df_2,on='Id',how='right').set_index('Id').sort_values('Id')
df_right

# df_right.count()
# df_right.isnull().sum(axis = 0)

# all rows of df_2 are included
```

```python
df_left= pd.merge(df_1,df_2,on='Id',how='left').set_index('Id').sort_values('Id')
df_left 

df_left.count()
# df_left.isnull().sum(axis = 0)

# all rows of df_1 are included
```

```python
df_concat1 = pd.concat([df_1,df_2], axis=1)
df_concat1

#stacking to the left side
```

```python
df_concat2 = pd.concat([df_1,df_2], axis=0)
df_concat2
 #stacking at the end
```

```python
df_1.join(df_2,on='Id',lsuffix='_1',rsuffix='_2')
```

# 4. Filter by Column Names

```python
name = ['Alex','Sam','Alice','John','Joseph']
age = ['34','45','25','30','40']
nationality= ['Nepalese','Indian','Chinese','Nepalese','American']
```

```python
ddct = {'Name':name,'Age':age,'Nationality':nationality}
```

```python
df_new = pd.DataFrame(ddct)
df_new
```

```python
#retrieving only name and age

df_new[['Name', 'Age']]
```

```python
#retrieving only name and age (using filter)
df_new.filter(["Name","Age"], axis=1)
```

```python
#filtering columns whose name starts with Na

filter_col = [col for col in df_new if col.startswith('Na')]
# filter_col
df_new[filter_col]
```

```python
#filtering columns whose name ends with e

filter_col2 = [col for col in df_new if col.endswith('e')]
# filter_col
df_new[filter_col2]
```

```python
# Retrieve all the data of the column having subset “am” 
df_new.iloc[:, lambda df_new: df_new.columns.str.contains('am',
                                              case=False)].head()
```

```python
#Retrieve all the data of the column having subset “am” (using filter)
# df_new.filter(regex='am')
df_new.filter(like='am')
```

```python
df_new.filter(like='4', axis=0) 
```

# 5. Countries COVID Data

```python
#cases by countries
df_link.groupby(by=["Country_Region"]).sum()
```

```python
# total active cases sum for Madhya Pradesh
total_mp = df_india.loc[df_india['Province_State'] == 'Madhya Pradesh', 'Active'].sum()
total_mp
```

```python
# total cases sum for Madhya Pradesh
df_india.loc[df_india['Province_State'] == 'Madhya Pradesh']
```

```python
# getting the state with maximum confirmed cases
# max_value = df_india['Confirmed'].max()
df_india.loc[df_india['Confirmed'] == df_india['Confirmed'].max()]
```

# Canada

```python
#cases for Canada
df_canada= df_link.loc[(df_link['Country_Region'] == 'Canada')]
df_canada
```

```python
#filling na values in Active column with median
df_canada['Active'].fillna(df_canada['Active'].median(), inplace=True)
print(df_canada['Active'])
```

# US Data

```python
#cases for US
df_us= df_link.loc[(df_link['Country_Region'] == 'US')]
df_us
```

```python
# cases for each states
df_us_states = df_us.groupby(['Province_State']).sum()
df_us_states
```

```python
# counting no of rows
len(df_us_states.index)
```

```python
#reseting index 
df_us_states.reset_index(inplace=True)
df_us_states = df_us_states.rename(columns = {'index':'Province_State'})
```

```python
df_us_states
```

```python
col_list = df_us_states['Province_State'].tolist()
sorted(col_list)
```

```python
not_states =['Diamond Princess','District of Columbia','Grand Princess',
             'Guam','Northern Mariana Islands','Puerto Rico','Recovered',
        'Virgin Islands',]
# res = set(col_list) - set(not_states)

df_us_states = df_us_states[df_us_states['Province_State'].isin(set(col_list) - set(not_states))]
df_us_states
```

```python
#state with minimum active case
df_us_states.loc[df_us_states['Active'] == df_us_states['Active'].min()]
```
