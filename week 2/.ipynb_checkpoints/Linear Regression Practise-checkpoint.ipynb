```python
#imports
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target
```

```python
# Code here    
X = pd.DataFrame(np.c_[boston['CRIM'], boston['ZN'],boston['INDUS'],boston['CHAS'],boston['NOX'],boston['RM'],
                       boston['AGE'],boston['DIS'],boston['RAD'],boston['TAX'],boston['PTRATIO'],boston['B']
                       ,boston['LSTAT']], columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 
                                                     'RAD', 'TAX','PTRATIO', 'B', 'LSTAT'])
Y = boston['MEDV']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# X_train

# Code here
scaler = StandardScaler()
X_train = scaler.fit_transform( X_train )
X_test = scaler.transform( X_test )
```

```python
W = np.zeros(X_train.shape[1]+1)
alpha =  0.1
W
```

```python
X_train_f = np.c_[np.ones(len(X_train),dtype='int64'),X_train]
# X_train_f
```

X_train in the form:
array([[1, 3],
       [1, 4],
       [1, 5]])                  
       1s in first column

```python
X_train_f.shape
```

```python

# Y_train = np.squeeze(np.asarray(Y_train))
Y_train
```

```python
# Solve using normal equation
W = np.linalg.inv(X_train_f.T.dot(X_train_f)).dot(X_train_f.T).dot(Y_train)
```

```python
W
```

```python
def predict(X, W):
    return X.dot(W)
#     return W'X
```

```python
# Code here
def mse(y, yhat):
    
    N = yhat.shape

    error = np.sum((yhat - y)**2)/(N)

    return error
```

```python
X_train_f.shape
```

```python
Y_train.shape
```

```python
yhat = predict(X_train_f, W)
# print(yhat)
# np.sum((yhat - Y_train)**2)/(yhat.shape)
```

```python
loss = mse(Y_train, yhat)
print(loss)
```

```python
def compute_gradient(X, y, W):
    # Code here
    
    m = X.shape[0]
    
    h = predict(X, W)
    loss = h - y
    
    # Gradient Calculation
    dW = X.T.dot(loss) / m
    
    return dW
```

```python
# Code here
W = W - alpha*compute_gradient(X_train_f, Y_train, W)
```

```python
W
```

```python
alpha = 0.1
N = 1000
losses = []
for i in range(N):
    
    # Predict
    yhat = predict(X_train_f, W)
    
    # Calculate Loss
    loss = mse(Y_train, yhat)
    
    # Compute gradient dL/dW
    dW = compute_gradient(X_train_f, Y_train, W)
    
    # Update the weights
    W = W - alpha*compute_gradient(X_train_f, Y_train, W)

    losses.append(loss)
```

```python
losses
```

```python
import matplotlib.pyplot as plt

x = np.arange(1000)
 
# plotting
plt.title("Line graph")
plt.xlabel("training iteration")
plt.ylabel("loss")
plt.plot(x,losses, color ="blue")
plt.show()
```
