Create two arrays: x, y  and z where:                                     
x contains 12 evenly spaced numbers between 1 and 20                                   
y contains the values of x multiplied by 2 ie. 2*x                   
z contains the squares of elements of x ie x**2   

Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)

Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.

Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

Finally, redo all the above plots using seaborn which is a higher level plotting library. It creates more aesthetically pleasing charts than matplotlib.


```python
#creating array containing 12 evenly spaced nos between 1 and 20
x = np.linspace(1,20,num=12)
x
```

```python
y = x*2
z= x**2
y
```

```python
# Plotting x vs y(2x)
plt.plot(x, y, 'r--')
plt.title(' y=2x ', fontdict={'fontsize': 20})
# X and Y labels
plt.xlabel('x',fontdict={'fontsize': 15})
plt.ylabel('y',fontdict={'fontsize': 15})
plt.show()
```

```python
plt.plot(x, y, 'blue')
plt.title(' y=2x with increased axis range', fontdict={'fontsize': 20})
plt.xlabel('x',fontdict={'fontsize': 15})
plt.ylabel('y',fontdict={'fontsize': 15})
plt.xlim(0, 25)
plt.ylim(0, 450)
plt.show()
```

```python
p1 = plt.plot(y, x, 'blue', label= 'y vs x: x = y/2')
func=r"$\sqrt{z}$"
p2 = plt.plot(z, x, 'purple', label = f'z vs x: x = {func}')
plt.title('y vs x and z vs x', fontdict={'fontsize': 20})
plt.legend()
plt.show()
```

# Subplots

```python
from matplotlib.pyplot import figure

fig = plt.figure(figsize =(12, 6))
plt.subplot(1, 2, 1)
plt.title('y vs x: x = y/2',fontdict={'fontsize': 20})
plt.xlabel('y ',fontdict={'fontsize': 15})
plt.ylabel('x ',fontdict={'fontsize': 15})
plt.plot(y,x)

plt.subplot(1, 2, 2)
func=r"$\sqrt{z}$"
plt.title(f'z vs y: y = 2{func}',fontdict={'fontsize': 20})
plt.xlabel('z ',fontdict={'fontsize': 15})
plt.ylabel('y ',fontdict={'fontsize': 15})
plt.plot(z,y)

plt.show()
```

```python
#plotting x vs y using seaborn
plt2 = sns.lineplot(x, y)
plt2.set(xlabel='x', ylabel='y')
plt2.set_title("y=2x",fontdict={'fontsize': 20})
plt.xlim(0, 25)
plt.ylim(0, 450)
plt.show()
```

```python
figure = plt.figure(figsize=(12,6))

axes1 = figure.add_subplot(121)
plt1 = sns.lineplot(y,x ,ax=axes1)
plt1.set(xlabel='y value', ylabel='x value')
plt1.set_title("y vs x : x = y/2 ",fontdict={'fontsize': 20})


axes2 = figure.add_subplot(122)
plt2 = sns.lineplot(z,y, ax=axes2)
plt2.set(xlabel='z', ylabel='y')
func=r"$\sqrt{z}$"
plt2.set_title(f"z vs y : y = 2{func}",fontdict={'fontsize': 20})
plt.show()
```

# CSV File

```python
df_pima= pd.read_csv('dataset/pima_diabetes.csv')  
df_pima.head()
```

# Bar Plot

```python
# counting no of 0s and 1s in outcome
df_pima['Outcome'].value_counts()
```

```python
df_outcome = pd.DataFrame({'Outcome':[df_pima['Outcome'].value_counts()[0],df_pima['Outcome'].value_counts()[1]], 'val':[0,1]})
df_outcome.plot.bar(x='val', y='Outcome', rot=0)
plt.title('Cout of occurance each class of Outcome variable')
plt.show()
```

The dataset is imbalanced as the no of data points of one class(class=0) is almost double the no of data points of another class(class =1). There are two classes of Outcome attribute: these classes are not represented equally. 

Here, 65% of the data belong to class 0 so even if we create Ml model that predicts class 0 for all input, machine will still have accuracy of 65%. Therefore, highly imbalanced data creates a problem while evaluating the model performance. Even model with high accuracy can give poor performance while evaluating on unseen data.
The classical performance measures used for evaluating the performances of classifiers when used with balanced datasets are not appropriate for imbalanced datasets.

# Histogram

```python
#distribution of data
fig = plt.figure(figsize =(10, 7))

for i in range(8):
    plt.subplot(2, 4, i+1) 
    plt.title(df_pima.columns[i])
    df_pima[df_pima.columns[i]].hist()

plt.subplots_adjust(left=0.000,
                    bottom=0.0, 
                    right=1, 
                    top=1, 
                    wspace=0.5, 
                    hspace=0.5)
plt.show()
```

Gulcose, Blood Pressure, BMI has nearly normal distribution. The histogram is bell-shaped, with only one peak, and is almost symmetric around the mean. 

Age, DiabetesPedigreeFunction, Pregnancies, Insulin, Skin Thickness have skewed(right-skewed) distribution

# Boxplot

```python
# increasing fig size
fig = plt.figure(figsize =(10, 7))

for i in range(8):
    plt.subplot(2, 4, i+1) 
    plt.title(df_pima.columns[i])
    plt.boxplot(df_pima.iloc[:,i:i+1])
    
plt.subplots_adjust(left=0.000,
                    bottom=0.0, 
                    right=1, 
                    top=1, 
                    wspace=0.5, 
                    hspace=0.5)
plt.show()
```

DiabetesPedigreeFunction, Age, BMI , Insulin, Blood Pressure have more outliers. There are more data points lying above or below whisker plot that is ouside of the range of (Q1-1.5IQR) and (Q1+1.5IQR)

Q1: 25th percentile      
Q2: 50th percentile  
Q3: 75th percentile  
IQR: Q3-Q1
