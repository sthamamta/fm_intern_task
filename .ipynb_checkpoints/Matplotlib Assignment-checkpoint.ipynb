```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

Create two arrays: x, y  and z where:                                     
x contains 12 evenly spaced numbers between 1 and 20                                   
y contains the values of x multiplied by 2 ie. 2*x                   
z contains the squares of elements of x ie x**2   

Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)

Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.

Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

Finally, redo all the above plots using seaborn which is a higher level plotting library. It creates more aesthetically pleasing charts than matplotlib.


```python
#creating array containing 12 evenly spaced nos between 1 and 20
x = np.linspace(1,20,num=12)
x
```

```python
y = x*2
z= x**2
y
```

```python
# Plotting x vs y(2x)
plt.plot(x, y, 'r--')
plt.title(' y=2x ', fontdict={'fontsize': 20})
# X and Y labels
plt.xlabel('x',fontdict={'fontsize': 15})
plt.ylabel('y',fontdict={'fontsize': 15})
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'purple')
plt.title(' y=2x ', fontdict={'fontsize': 20})
plt.xlabel('x',fontdict={'fontsize': 15})
plt.ylabel('y',fontdict={'fontsize': 15})
plt.show()
```

```python
plt.plot(x, y, 'blue')
plt.title(' y=2x with increased axis range', fontdict={'fontsize': 20})
plt.xlabel('x',fontdict={'fontsize': 15})
plt.ylabel('y',fontdict={'fontsize': 15})
plt.xlim(0, 25)
plt.ylim(0, 450)
plt.show()
```

```python
p1 = plt.plot(y, x, 'blue', label= 'y vs x')
p2 = plt.plot(z, x, 'purple', label = 'z vs x')
plt.title('y vs x and z vs x', fontdict={'fontsize': 20})
plt.legend()
plt.show()
```

# Subplots

```python
from matplotlib.pyplot import figure

fig = plt.figure(figsize =(12, 6))
plt.subplot(1, 2, 1)
plt.title('y vs x: x = y/2',fontdict={'fontsize': 20})
plt.xlabel('y ',fontdict={'fontsize': 15})
plt.ylabel('x ',fontdict={'fontsize': 15})
plt.plot(y,x)

plt.subplot(1, 2, 2)
plt.title('z vs y: y = sqrt(z)*2',fontdict={'fontsize': 20})
plt.xlabel('z ',fontdict={'fontsize': 15})
plt.ylabel('y ',fontdict={'fontsize': 15})
plt.plot(z,y)

plt.show()
```

# Seaborn

```python
#plotting x vs y using seaborn
plt1 = sns.lineplot(x, y)
plt1.set(xlabel='x', ylabel='y')
plt1.set_title("y=2x",fontdict={'fontsize': 20})
plt.show()
```

```python
#plotting x vs y using seaborn
plt2 = sns.lineplot(x, y)
plt2.set(xlabel='x', ylabel='y')
plt2.set_title("y=2x",fontdict={'fontsize': 20})
plt.xlim(0, 25)
plt.ylim(0, 450)
plt.show()
```

```python
sns.lineplot(y,x ,label="y vs x : x= y/2")  #horizontal axis=>y vertical axis=>x
sns.lineplot(z,x ,label="z vs x : x= sqrt(z)")  #horizontal axis=>z vertical axis=>x
plt.show()
```

```python
figure = plt.figure(figsize=(12,6))

axes1 = figure.add_subplot(121)
plt1 = sns.lineplot(y,x ,ax=axes1)
plt1.set(xlabel='y value', ylabel='x value')
plt1.set_title("y vs x : x = y/2 ",fontdict={'fontsize': 20})


axes2 = figure.add_subplot(122)
plt2 = sns.lineplot(z,y, ax=axes2)
plt2.set(xlabel='z', ylabel='y')
plt2.set_title("z vs y : y = sqrt(z)*2",fontdict={'fontsize': 20})
plt.show()
```

Plot the scatter plot of x1 and x2. Color the points in the scatterplot according to their values in y: blue for 0 and red for 1. For example, color the point 
(5, 99) blue as the corresponding y value is 0 and the point (7, 86) red as the corresponding y value is 1.  


```python
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])

x2 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

y = np.array([0, 1, 1, 0, 0, 1, 0, 0, 1, 0, 0, 1, 1])

plt.scatter(x=x1, y=x2, c=['black' if i  == 0 else 'red' for i in y])
plt.show()
```

# CSV File

```python
df_pima= pd.read_csv('dataset/pima_diabetes.csv')  
df_pima.head()
```

# Bar Plot

```python
# counting no of 0s and 1s in outcome
df_pima['Outcome'].value_counts()
```

```python
df_outcome = pd.DataFrame({'Outcome':[df_pima['Outcome'].value_counts()[0],df_pima['Outcome'].value_counts()[1]], 'val':[0,1]})
df_outcome.plot.bar(x='val', y='Outcome', rot=0)
plt.title('Cout of occurance each class of Outcome variable')
plt.show()
```

The dataset is imbalanced as the no of data points of one class is double the no of data points of another class

```python
#checking data distribution of single column
df_pima['Pregnancies'].hist(bins=12, alpha=0.8)
plt.show()
```

# Histogram

```python
#distribution of data
fig = plt.figure(figsize =(10, 7))

# plot 1
plt.subplot(2, 4, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.title('BloodPressure')
df_pima['BloodPressure'].hist(bins=12, alpha=0.8)

#plot 2
plt.subplot(2, 4, 2) 
plt.title('Pregnancies')
df_pima['Pregnancies'].hist(bins=12, alpha=0.8)

#plot 3
plt.subplot(2, 4, 3)
plt.title('Glucose')
df_pima['Glucose'].hist(bins=12, alpha=0.8)

#plot 4
plt.subplot(2, 4, 4)
plt.title('SkinThickness')
df_pima['SkinThickness'].hist(bins=12, alpha=0.8)

#plot 5
plt.subplot(2, 4, 5)
plt.title('Insulin')
df_pima['Insulin'].hist(bins=12, alpha=0.8)

#plot 6
plt.subplot(2, 4, 6)
plt.title('BMI')
df_pima['BMI'].hist(bins=12, alpha=0.8)

#plot 7
plt.subplot(2, 4, 7)
plt.title('Age')
df_pima['Age'].hist(bins=12, alpha=0.8)

#plot 8
plt.subplot(2, 4, 8)
plt.title('DiabetesPedigreeFunction')
df_pima['DiabetesPedigreeFunction'].hist(bins=12, alpha=0.8)

plt.subplots_adjust(left=0.000,
                    bottom=0.0, 
                    right=1, 
                    top=1, 
                    wspace=0.5, 
                    hspace=0.5)
plt.show()
```

age, diabetesPedigreeFunction, Pregnancies, Insulin these data have skewed distribution

# Boxplot

```python
# increasing fig size
fig = plt.figure(figsize =(10, 7))

# plot 1
plt.subplot(2, 4, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.title('BloodPressure')
plt.boxplot(df_pima['BloodPressure'])

#plot 2
plt.subplot(2, 4, 2) 
plt.title('Pregnancies')
plt.boxplot(df_pima['Pregnancies'])

#plot 3
plt.subplot(2, 4, 3)
plt.title('Glucose')
plt.boxplot(df_pima['Glucose'])

#plot 4
plt.subplot(2, 4, 4)
plt.title('SkinThickness')
plt.boxplot(df_pima['SkinThickness'])

#plot 5
plt.subplot(2, 4, 5)
plt.title('Insulin')
plt.boxplot(df_pima['Insulin'])

#plot 6
plt.subplot(2, 4, 6)
plt.title('BMI')
plt.boxplot(df_pima['BMI'])

#plot 7
plt.subplot(2, 4, 7)
plt.title('Age')
plt.boxplot(df_pima['Age'])

#plot 8
plt.subplot(2, 4, 8)
plt.title('DiabetesPedigreeFunction')
plt.boxplot(df_pima['DiabetesPedigreeFunction'])

plt.subplots_adjust(left=0.000,
                    bottom=0.0, 
                    right=1, 
                    top=1, 
                    wspace=0.5, 
                    hspace=0.5)
plt.show()
```

DiabetesPedigreeFunction, Age, BMI , Insulin, Blood Pressure have more outliers

```python
figure = plt.figure(figsize=(15,6))

axes = figure.add_subplot(111)

corr = df_pima.corr()
corr_mat = axes.matshow(corr)

plt.colorbar(corr_mat)

axes.set_yticklabels(list(df_pima.columns))

plt.show()
```
